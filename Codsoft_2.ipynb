```python
import datetime

print("=" * 60)
print("        RULE-BASED CHATBOT")
print("=" * 60)
print("Type 'exit' or 'bye' anytime to quit.")
print()

name = input("Enter your name: ")

print(f"\nHello {name}! I am your Rule-Based Chatbot.")
print("How can I help you today?\n")
```

```python
def calculator():
    try:
        num1 = float(input("Enter First Number: "))
        op = input("Enter Operator (+ - * /): ")
        num2 = float(input("Enter Second Number: "))

        if op == "+":
            print("Result =", num1 + num2)

        elif op == "-":
            print("Result =", num1 - num2)

        elif op == "*":
            print("Result =", num1 * num2)

        elif op == "/":
            if num2 == 0:
                print("Cannot divide by zero.")
            else:
                print("Result =", num1 / num2)

        else:
            print("Invalid Operator.")

    except:
        print("Invalid Input.")


while True:

    user = input("\nYou : ").lower()

    # Exit
    if user in ["exit", "bye", "quit", "goodbye"]:
        print("Bot : Goodbye! Have a nice day.")
        break

    # Greetings
    elif "hello" in user or "hi" in user or "hey" in user:
        print("Bot : Hello! Nice to meet you.")

    # Asking Name
    elif "your name" in user:
        print("Bot : My name is RuleBot.")

    # User Name
    elif "my name" in user:
        print(f"Bot : Nice to know you, {name}_!")

    # How are you
    elif "how are you ?" in user:
        print("Bot : I'm doing great. Thank you for asking!")

    # Who created you
    elif "who created you ? " in user or "developer" in user:
        print("Bot : I was created using Python as a Rule-Based Chatbot project.")

    # Time
    elif "time ?" in user:
        current = datetime.datetime.now()
        print("Bot :", current.strftime("%I:%M:%S %p"))

    # Date
    elif "date ?" in user:
        current = datetime.datetime.now()
        print("Bot :", current.strftime("%d-%m-%Y"))

    # Day
    elif "day ?" in user:
        current = datetime.datetime.now()
        print("Bot :", current.strftime("%A"))

    # Weather
    elif "weather ?" in user:
        print("Bot : Sorry, I cannot access live weather.")
        print("Bot : Please check a weather website.")

    # Joke
    elif "joke" in user:
        print("Bot : Why do programmers prefer Python?")
        print("Bot : Because it's easy to understand and powerful!")

    # Thanks
    elif "thank" in user:
        print("Bot : You're welcome!")

    # Help
    elif "help" in user:
        print("\nI can answer questions like:")
        print("- hello")
        print("- how are you")
        print("- your name")
        print("- date")
        print("- time")
        print("- day")
        print("- joke")
        print("- weather")
        print("- calculator")
        print("- bye")

    # Calculator
    elif "calculator" in user or "calculate" in user or "math" in user:
        calculator()

    # Favourite language
    elif "favorite language ?" in user:
        print("Bot : Python is my favorite language!")

    # AI
    elif "ai" in user:
        print("Bot : AI stands for Artificial Intelligence.")

    # Python
    elif "python" in user:
        print("Bot : Python is a simple and powerful programming language.")

    # College
    elif "college ?" in user:
        print("Bot : I don't attend college, but I love helping students.")

    # Age
    elif "age" in user:
        print("Bot : I don't have an age like humans.")

    # Default Response
    else:
        print("Bot : Sorry, I don't understand that.")
        print("Bot : Type 'help' to see what I can do.")
```
